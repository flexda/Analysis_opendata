### 연도와 월 분리하기 

* data.split()
* data.split()[].replace()

* parse_year() 사용자 정의 함수 만들기   
* parse_month()   

* apply()

```python
date="2013년 12월"
```

```python
date.split()
```

```python
date.split()[1]
```

```python
date.split()[0]
```

```python
date.split("년")
```

```python
date.split("년")[-1].replace("월","")
```

```python
def parse_year(date):
    year=date.split("년")[0]
    year=int(year)
    return year
```

```python
parse_year(date)
```

```python
type(parse_year(date))
```

```python
def parse_month(date):
    month=date.split("년")[-1].replace("월","")
    month=int(month)
    return month
```

```python
parse_month(date)
```

```python
df_first_melt["연도"]=df_first_melt["기간"].apply(parse_year)
```

```python
df_first_melt.head(1)
```

```python
df_first_melt["월"]=df_first_melt["기간"].apply(parse_month)
```
